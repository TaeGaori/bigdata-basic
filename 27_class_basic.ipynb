# 클래스(class)와 객체 지향 프로그래밍(OOP)

## 1. 클래스란 무엇인가???

**클래스(class)**는 데이터와 그 데이트를 다루는 함수(메서드)를 하나로 묶어 새로운 타입을 정의하는 방법(==설계도)

|        용어       |        설명           |             비유               |
|-----|------|------|
| 클래스(class) | 설계도 | 붕어빵 틀, 와플 틀, 과자 틀 |
|인스턴스(instance) | 클래스로 만든 살제 객체 | 붕어빵 한 개, 와플 1개, 과자 1개 |
|메서드(method) | 클래스 내부에서 만들어진 함수 | 붕어빵의 기능, 와플 기능, 과자 기능 |
|속성(attribute) | 인스턴스가 가진 데이터 (==인스턴스 변수) | 붕어빵의 재료(맛)

## 2. 클래스 기본 구조

```python
# 클래스 정의의 기본 구조
# class 클래스이름 : 로 시작한다.
# 파이썬에서는 클래스이름 첫글자는 대무자로 하는 것이 관행이다.
class Toy:
    # 메서드(함수) 정의
    # __init__ : 객체(인스턴스)가 처음 만들어질 때 자동으로 호출되는 '초기화 메서드' (생성자)
    # self : 현재 만들어진 인스턴스(객체) 자기 자신을 가리킨다.
    def __init__(self):
        self.elems = []  # self.속성 --> 인스턴스 변수 : 각 객체가 개별로 가지는 데이터

    # 메서드(함수) 정의 --> 인스턴스 메서드 정의
    def add(self, new_elems):
        """
        첫번째 매개변수 반드시 self --> 호출한 인스턴스가 자동으로 전달된다.
        new_elems 매개변수는 리스트이다.
        """
        self.elems += new_elems    #new_elems만큼 누적해라!

    def size(self):
        """
        저장된 요소 개수를 반환(리턴)한다.
        개수를 우리에게 알려준다.
        """
        return len(self.elems)
```

## 3. 클래스로 인스턴스(객체) 만들기

> 인스턴스이름 = 클래스이름()

__init__ 자동으로 실행되면서 새 객체가 생성된다.

```python
t1 = Toy()  # t1과 t2는 서로 다른 독립적인 객체이다
t2 = Toy()

print(t1)
print(t2)
```

```python
t1 is t2 # 같은 클래스로 만들었지만 서로 다른 객체다!
```

## 4. 메서드 호출

## 간단한 클래스 예제

```python
class Person:

    def __init__(self, name, age):
        self.name = name 
        self.age = age

    def print_into(self):
        print(f'{self.name}은(는) {self.age}살입니다!')
        
```

```python
saram1 = Person('홍길동', 30)
saram2 = Person('임꺽정', 40)
saram3 = Person('이순신', 50)
saram4 = Person('정약용', 60)
```

```python
print(saram1)
print(saram2)
print(saram3)
print(saram4)
```

```python
saram1.print_into()
saram2.print_into()
saram3.print_into()
saram4.print_into()
```

### 돌발퀴즈!
- Person 클래스에 인스턴스 변수(속성)추가
    -직업, 취미

- Person 클래스에 매서드를 추가해보자!
    -ex) 홍길동의 직업은 나무꾼입니다!   -> 각각
    -ex) 홍길동의 취미는 요리하기 입니다!

```python
class Person:

    def __init__(self, name, age, job, hobby):
        self.name = name 
        self.age = age
        self.job = job
        self.hobby = hobby

    def print_into(self):
        print(f'{self.name}은(는) {self.age}살입니다!')

    def print_job(self):
        print(f'{self.name}의 직업은 {self.job}입니다!')
        
    def print_hobby(self):
        print(f'{self.name}의 취미는 {self.hobby}입니다!')
```

```python
saram5 = Person('홍길순', 20, '나무꾼', '요리하기')
saram5.print_job()
saram5.print_hobby()
```
