# PT - 4

```python

titanic = pd.read_csv('train.csv')
titanic
```

```python
ti1 = titanic[['Sex', 'Survived']]
ti1
```

```python
ti1.groupby('Sex').mean()
```

```python
ti2 = titanic[['Parch', 'Survived']]
ti2
```

```python
ti2.groupby('Parch', as_index=False).mean().sort_values('Survived', ascending = False)
```

```python
ti3 = titanic[['Pclass', 'Sex', 'Survived', 'Age']]
ti3
```

```python
# 데이터프레임명.groupby([그룹할 컬럼1, 그룹할 컬럼2])[[계산할 컬럼1, 계산할 컬럼2]].mean()

ti3.groupby(['Pclass', 'Sex']).mean()
```

```python
titanic.groupby(['Pclass', 'Sex'])[['Survived', 'Age']].mean()
```

```python
ti4 = titanic[['Embarked', 'Survived']]
ti4
```

```python
ti4.plot(kind = 'line', x = 'Embarked', y = 'Survived')
plt.show()
```

```python
# seaborn 사용 --> 생존자 --> 건수 --> countplot()
sns.countplot(data = ti4, x = 'Embarked', hue = 'Survived') #데이터의 특정 컬럼 값을 기준으로 그룹을 나누어 색상으로 구분하여 표시하는 파라미터
plt.show()
```
