```python
# 1.수집

url = "https://api.odcloud.kr/api/15059963/v1/uddi:76bba8dc-16b6-4898-96af-e3c056854ec3"

params = {
    "serviceKey": API_KEY
}

response = requests.get(url, params=params)

print("응답 코드:", response.status_code)
```

```python
# 2. 저장
params = {
    "serviceKey": API_KEY,
    "page": 1,
    "perPage": 1000
}

response = requests.get(url , params=params)

if response.status_code == 200:
    raw_data = response.json()
    items = raw_data.get("data", [])
    df = pd.DataFrame(items)
    csv_filename = "lotto_winners.csv"
    df.to_csv(csv_filename, index=False, encoding="utf-8-sig")
    print(f"[저장 완료] 원본 데이터를 '{csv_filename}' 파일로 보관했습니다.")
```
