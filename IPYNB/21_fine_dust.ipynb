# 에어코리아 - 미세먼지 (2001년 ~ 2019년)

```python
## fine_dust.xlxs 파일 불러오기 & 라이브러리 불러오기
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

data = pd.read_excel('fine_dust.xlsx')  #인덱스 행이 나온다   인덱스 : 첫 행에 숫자 나옴.
data
```

```python
data = pd.read_excel('fine_dust.xlsx', index_col = 'area')  #인덱스 행 --> area로
data
```

## 2018년 지역별 미세먼저 선 그래프

```python
data2018 = data[2018]
data2018
```

```python
plt.figure(figsize=(15, 4)) #가로 15인치, 세로 4인치
plt.plot(data2018, color='purple', marker='o')
plt.title('2018년 미세먼지 선 그래프')
plt.xlabel('지역')
plt.ylabel('micrometer')
plt.grid()  #격자무늬
plt.show()
```

## 2016 ~ 2019 지역별 미세먼지 선 그래프

```python
for year in range(2016, 2020):
    chartdata = data[year]
    print(chartdata)
    print('-' * 50)
```

```python
plt.figure(figsize=(15, 4))
for year in range(2016, 2020):
    chartdata = data[year]
    plt.plot(chartdata, marker='s', label=year)
    
plt.title('2016 ~ 2019 미세먼지 선 그래프')
plt.xlabel('지역')
plt.ylabel('micrometer')
plt.legend
plt.grid()
plt.show()
```

## 2017년 지역별 미세먼지 세로 막대 그래프

- plt.bar(x = x의 데이터, height = y의 데이터)

```python
data2017 =data[2017]
data2017
```

```python
plt.figure(figsize=(15, 4))
plt.bar(data2017.index, data2017, color='g')
plt.title('2017년 지역별 미세먼지')
plt.xlabel('지역')
plt.ylabel('micrometer')
plt.legend
plt.grid()
plt.show()
```

## 2016 ~ 2019 경기, 인천, 대구 지역별 미세먼지 그룹 세로 막대 그래프

```python
data = pd.read_excel('fine_dust.xlsx', index_col='area')
data
```

```python
data4 = data.loc['Gyeonggi': 'Daegu']
data4
```

```python
data4 = data4.loc[:, 2016:2019]
data4
```

```python
plt.figure(figsize=(15, 4))
for year in range(2016, 2020):
    chartdata = data4[year]
    plt.bar(data4.index, chartdata, label = year)

plt.show()
```

```python
import numpy as np
index = np.arange(4)    # 0~3

plt.figure(figsize=(15, 4))
for year in range(2016, 2020):
    chartdata = data4[year]
    plt.bar(index, chartdata,width = 0.2, label = year)
    index = index + 0.2

plt.ylim(35, 55)
plt.xlabel('지역')
plt.ylabel('micrometer')
plt.title('2016~2019 미세먼지 그룹별 세로 막대 그래프')     #경기, 인천, 부산, 대구
plt.xticks(index-0.5, ['Gyeonggi', 'Incheon', 'Busan', 'Daegu'])
plt.legend()
plt.show()
```

## 2016 ~ 2019 지역별 미세먼지 그룹 누적 가로 막대 그래프

```python
plt.figure(figsize=(10, 4))

data4 = data.loc['Gyeonggi' : 'Daegu', 2016:2019]
for year in range(2016,2020):  
    plt.barh(data4.index, data4[year], label = year)

plt.xlim(35, 55)
plt.ylabel('지역')
plt.xlabel('micrometer')
plt.title('2016~2019 지역별 미세먼지 그룹 누적 가로그래프')     #경기, 인천, 부산, 대구
plt.yticks(index-0.5, ['Gyeonggi', 'Incheon', 'Busan', 'Daegu'])
plt.legend()
plt.show()
```
