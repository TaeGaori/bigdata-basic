### 조건에 맞는 데이터 추출

```python
#최고로 더웠떤 날의 모든 정보(행)
w = df['temp'] == df['temp'].max()
df[w]
```

## 조건이 2개 이상 --> 그리고 &, 또는 |

```python
# 기온이 30도 이상이고, 최대풍속이 9이상인 데이터

w = (df['temp'] >= 30) & (df['max_wind'] >= 9)
df[w]
```

```python
# 기온이 30도 이상이거나 , 최대풍속이 9이상인 데이터

w = (df['temp'] >= 30) | (df['max_wind'] >= 9)
df[w]
```

## 결측데이터 대체하기 --> fillna(채울 값)

```python
df['max_wind'].isnull().sum()
```

```python
df3 = df.copy() #복사
df3
```

```python
df3.fillna(0)
```

```python
df3 = df3.fillna(0)
```

```python
df3.isnull().sum()
```

```python
# 평균 --> mean()
avg = df3['max_wind'].mean()
avg
```

```python
df3['max_wind'] = df3['max_wind'].fillna(avg)
```

```python
avg2 = df3['mean_wind'].mean()
avg2
```

```python
df3['mean_wind'] = df3['mean_wind'].fillna(avg2)
```
