```python
# 라이브러리 불러오기

import pandas as pd
import numpy as np
```

```python
dust = pd.read_excel('dust1.xlsx')
dust.head()
```

```python
# 컬럼 이름 변경

dust = dust.rename(columns={'측정일시' : 'date'})
dust.head()
```

```python
# 문자열로 형 변환

dust['date'] = dust['date'].astype(str)
dust['date'] = dust['date'].str.slice(0, 8)
dust.dtypes
```

```python
# 날짜형으로 형 변환

dust['date'] = pd.to_datetime(dust['date'])
dust.dtypes
```

# 날짜형을 년, 월, 일로 분리 추출

```python
# dt 속성

dust['date'].dt
```

```python
dust['date'].dt.year    #연도 추출
```

## 년, 월, 일 컬럼을 생성

```python
dust['year'] = dust['date'].dt.year #연도 추출
dust['month'] = dust['date'].dt.month   #월 추출
dust['day'] = dust['date'].dt.day   #일 추출
dust.head()
```

# 데이터 병합 

```python
s1 = pd.read_excel('nation.xlsx')
s2 = pd.read_excel('code.xlsx')
s1.head(2)
```

```python
s2.head(2)
```

## pd.merge() 함수

- pd.merge(데이터프레임1, 데이터프레임2, on = 기준컬럼명, how = 'inner')
- how에는 'outer', 'left', 'right' 도 가능

```python
# 국적 코드 컬럼 기준

pd.merge(s1, s2, on = '국적코드')
```

```python
# 오른쪽 컬럼 기준 (ex.차집합)

pd.merge(s1, s2, how = 'right', on = '국적코드')
```

# 데이터 그룹핑

## groupby()

```python
# 국적코드별 성별, 입국객수, 증가수
s1.groupby('국적코드').sum()
```

```python
#국적코드별, 성별, --> 입국객수, 증가수
#그룹을 2개
s1.groupby(['국적코드', '성별']).sum()
```
