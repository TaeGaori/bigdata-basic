## 데이터프레임(DataFrame) 만들기

### 넘파이(numpy)로 배열 생성

```python
import numpy as np

arr = np.arange(12).reshape(4, 3)   # 행 4 , 열 3
arr
```

```python
print(arr)
```

```python
# 넘파이 배열을 데이터프레임으로
df = pd.DataFrame(arr)
df
```

## 데이터를 크기순으로 정렬 - sort_values()

- ascendint = True (생략) --> 오름차순
- ascendint = False --> 내림차순
- by = [컬럼명]

```python
# 최대 풍속 (max_wind) 데이터 크기에 따라 오름차순으로 정렬

df.sort_values(by = 'max_wind', ascending = True)
```

```python
df.sort_values(by = 'max_wind', ascending = False)
```

## 빈도 분석 결과 - value_counts()

```python
filename = 'bank.csv'
bank = pd.read_csv(filename)
bank.head()
```

```python
bank['job'].value_counts()  #내림차순 기본
```

```python
bank['job'].value_counts(ascending = True)  #오름차순
```
