# 대구광역시 달서구 cctv 개수 현황

```python
import csv

with open('cctv_20250527.csv', 'rt') as f:
    buffer = csv.reader(f, delimiter=',')
    totalcctv = 0 # cctv 개수
    for i, line in enumerate(buffer):
        if i != 0:
            totalcctv += int(line[6])
```
