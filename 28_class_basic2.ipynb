# 클래스와 클래스 변수

## 클래스 변수

- 여러 인스턴스가 변수공간을 같이 사용하는 변수
- 같은 클래스를 통해서 만들어진 인스턴스들끼리 공간을 사용

```python
class Person:
    population = 0  # 클래스 변수 (인구 수 저장)

    def __init__(self, name):
        """생성자 : 인스턴스가 생성되면 인구수 1명 증가, 이름 부여"""
        self.name = name # 인스턴스 변수(self.name)
        Person.population += 1  # 인구수 1 증가

    def print_hobby(self, hobby):
        """가지고 잇는 취미를 알려주는 인스턴스 매서드"""
        print(f'{self.name}이(가) {hobby}을(를) 즐깁니다!')
```

```python
saram1 = Person('라이언')   # 인스턴스(객체) 생성
saram2 = Person('춘식이')
saram3 = Person('무지')
saram4 = Person('어차피')
saram5 = Person('죠르디')
```

```python
saram1.print_hobby('뛰어놀기')
saram2.print_hobby('라이언과 놀기')
saram3.print_hobby('콘과 놀기')
saram4.print_hobby('장난치기')
saram5.print_hobby('뒹굴뒹굴')
```

```python
print(f'생성된 사람들은 모두 {Person.population}명 입니다!')
```

```python
saram6 = Person('네오')
saram6.print_hobby('꽃단장하기')
```

```python
print(f'생성된 사람들은 모두 {Person.population}명 입니다!')
```

## 클래스 매서드(class method)

- 인스턴스(객체)가 아니라 클래스 자체를 대상으로 동작하는 메서드
- 클래스 변수 수정
- 클래스 전체 설정 변경
- @classmethod라는 데코레이터 사용
- 첫번째 매개변수로 cls 사용
- 클래스 매서드는 인스턴스(객체)를 만들지 않고 바로 클래스명.클래스메서드()로 호출 가능

### 왜 사용하는가?

- 보통 여러 객체(인스턴스)가 있을 때 공통 데이터를 관리하고자 할 때 사용
- ex) 학교이름, 회사이름, 환율설정, 게임서버상태, 환경설정 값 등

```python
class Student:
    school = '석쌤학교' #클래스 함수

    def instance_method(self):
        print('self가 있으니 인스턴스 메서드')

    @classmethod
    def class_method(cls):
        print('')
```

```python
def instance_method(self):
        print('self가 있으니 인스턴스 메서드')

@classmethod
def class_method(cls):
    print('cls가 있으니 클래스 메서드')
```

```python
Student.class_method()
```

```python
s1 = Student()
s1.instance_method()
```

```python
# 수정 버전
class Person:
    population = 0  # 클래스 변수 (인구 수 저장)

    def __init__(self, name):
        """생성자 : 인스턴스가 생성되면 인구수 1명 증가, 이름 부여"""
        self.name = name # 인스턴스 변수(self.name)
        Person.population += 1  # 인구수 1 증가

    def print_hobby(self, hobby):
        """가지고 잇는 취미를 알려주는 인스턴스 매서드"""
        print(f'{self.name}이(가) {hobby}을(를) 즐깁니다!')

    @classmethod
    def get_population(cls):
        """인구수를 반환해주는 클래스 메서드"""
        return cls.population
```

```python
Person.get_population()
```

```python
print(f'현재 인구 수는 {Person.get_population()}명입니다!')
```

```python
saram1 = Person('라이언')   # 인스턴스(객체) 생성
saram2 = Person('춘식이')
saram3 = Person('무지')
```

### 실습 예제 - 분식집 -> 메뉴, 개수 -> 총 금액

```python
class Korean_snack:
    total = 0   # 총 금액 (클래스 변수)

    menu_list = [
        # 딕셔너리 {'음식이름' : 가격}
        {'떡볶이' : 3000},
        {'순대' : 3000},
        {'튀김' : 500},
        {'김밥' : 2000}
    ]   # 메뉴판 (클래스 변수)

    @classmethod
    def sales(cls, food, count):
        """
        총 계산 클래스 메서드
        
        매개변수(파라미터)
        food : 음식이름
        count: 개수
        """

        for menu in cls.menu_list:
            if food in menu:
                cls.total += (menu[food] * count)
```

```python
Korean_snack.sales('떡볶이', 2)
Korean_snack.sales('순대', 1)
Korean_snack.sales('튀김', 5)
```

```python
print(f'분식집에 내야 할 금액 : {Korean_snack.total}원')
```
