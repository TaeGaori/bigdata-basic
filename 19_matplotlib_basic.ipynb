# matplotlib 라이브러리 기본 시각화

```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
```

```python
# 데이터
xdata = [2, 3, 4, 5]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata)  # 선 그래프 (라인 그래프, 꺾은선 그래프)
plt.title('Line Graph')
plt.show()
```

## 범례 -->plt.legend()

```python
data1 = [2, 3, 4 ,5]
data2 = [8, 6, 4, 2]
plt.plot(data1, label = 'asc')     # asc -> 오름차순
plt.plot(data2, label = 'desc')    # desc -> 내림차순
plt.legend()
plt.show()
```

## 선에 색깔 넣기

```python
plt.plot(xdata, ydata, color = 'red')
plt.show()
```

## x축 제목, y축 제목

```python
plt.plot(xdata, ydata, color='c')
plt.xlabel('x축 제목')
plt.ylabel('y축 제목')
plt.show()
```

## 선 종류

- linestyle = 'dashed' or linestyle = '--'
- 일반 실선 linestyle = 'solid' or 'linestyle'

```python
plt.plot(data1, color = 'r', label = 'dashed', linestyle = '--') # 대시 긴 점선
plt.plot(data2, color = 'g', label = 'dotted', linestyle = ':')   #점선
plt.plot(xdata, color = 'b', label = 'dash-dot', linestyle = '-.') # 1점쇄선
plt.legend()
plt.show()
```

## Marker(마커) --> 점 모양

```python
plt.title('X, Y range')
xdata = [5, 10, 15, 20]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color='purple', linestyle='--', marker='o', markersize='10')
plt.show()
```

## 축 범위 조정
- plt.xlim(범위)
- plt.ylim(범위)

```python
plt.title('X, Y range')
xdata = [5, 10, 15, 20]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color='purple', linestyle='--', marker='o', markersize='10')
plt.xlim(0, 25)
plt.ylim(-2, 10)
plt.show()
```
